```python
import pandas as pd
```

```python
dataset = pd.read_csv('/content/drive/My Drive/캡스톤 디자인/csv/wadiz_3.csv')
```

```python
dataset = dataset.dropna()
```

```python
dataset.head()
```

```python
del dataset['Unnamed: 0']
dataset.columns = ['title','summary','image','gif','video','price','target_price','facebook','new','likes','feedback','search_word','supporter','novelty','noun','main','main_length','success','period']
dataset.head(10)
```

# Train Test Set

```python
X = dataset[['image','gif','video','price','target_price','period','facebook','new','likes','feedback','supporter','main_length',]]
y = dataset['success']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

# 정규화

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# **다중공선성 및 상관계수**

```python
from statsmodels.formula.api import ols
```

```python
model = ols('success ~ image + gif + video + price + target_price + period + facebook + new + likes + feedback + supporter + main_length', dataset)
```

```python
res = model.fit()
```

```python
res.summary()
```

다중공선성 확인

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
model.exog_names
```

```python
pd.DataFrame({'컬럼': column, 'VIF': variance_inflation_factor(model.exog, i)} 
             for i, column in enumerate(model.exog_names)
             if column != 'Intercept')  # 절편의 VIF는 구하지 않는다.
```

```python
feature = ['image','gif','video','price','target_price','period','facebook','new','likes','feedback','supporter','main_length']
```

```python
def model_coef(model):
  model_df = pd.DataFrame()
  coef = []
  for i in model:
    for j in i:
       coef.append(j)

  model_df['X'] = feature
  model_df['coef'] = coef

  return model_df.sort_values('coef',ascending =False)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
LI = LinearRegression()
LI.fit(X_train, y_train)
```

```python
print('train_set : ',LI.score(X_train, y_train))
print('test_set : ',LI.score(X_test, y_test))
```

```python
LI_df = pd.DataFrame()
linearcoef = []
for i in LI.coef_:
  linearcoef.append(i)
LI_df['X'] = feature
LI_df['coef'] = linearcoef
LI_df.sort_values('coef',ascending =False)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, y_train)
```

```python
print('train_set : ',LR.score(X_train, y_train))
print('test_set : ',LR.score(X_test, y_test))
```

Logistic Regression 상관계수

```python
model_coef(LR.coef_)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(X_train, y_train)
```

```python
print('train_set : ',DT.score(X_train, y_train))
print('test_set : ',DT.score(X_test, y_test))
```
