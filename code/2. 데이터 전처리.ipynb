```python
import pandas as pd
import re
```

```python
dir_link = '/content/drive/MyDrive/캡스톤 디자인/csv'
```

```python
dataset = pd.read_csv(dir_link+'/wadiz.csv')
dataset.head()
```

가격

```python
price1 = dataset['가격']
price = []
for i in price1:
  j = i.split('원')
  list = []
  
  j[0] = int(j[0].replace(",",""))
  price.append(j[0])

del dataset['가격']
dataset['가격'] = price
```

목표 금액

```python
cha_price1 = dataset['목표 금액']
cha_price = []
for i in cha_price1:
  j = i[5:]
  j = j.replace('원','')
  j = j.replace(',','')
  j = int(j)
  cha_price.append(j)
del dataset['목표 금액']
dataset['목표 가격'] = cha_price
```

펀딩 기간

```python
from datetime import datetime


period1 = dataset['펀딩 기간']
period = []
for i in period1:
  j = i.split('펀딩 기간 ')
  j = j[1]
  j1 = j[0:10]
  j1 = datetime.strptime(j1, '%Y.%m.%d')
  j2 = j[11:21]
  j2 = datetime.strptime(j2, '%Y.%m.%d')


  date = j2-j1
  date1 = str(date).split(' days')
  date = date1[0]
  period.append(date)
del dataset['펀딩 기간']
dataset['펀딩 기간'] = period
```

```python
fb1 = dataset['페이스북 지지자 수']
fb = []
for i in fb1:
  if i == ' ':
    i = '0'
    fb.append(int(i))
  else:
    fb.append(int(i))

del dataset['페이스북 지지자 수']
dataset['페이스북 지지자 수'] = fb
```

```python
new1 = dataset['새소식 수']
new = []
for i in new1:
  i = i[4:]
  i = i.replace('','0')
  new.append(int(i))

del dataset['새소식 수']
dataset['새소식 수'] = new
```

```python
supporter1 = dataset['서포터 수']
supporter = []
for i in supporter1:
  i = i[4:]
  if i == '':
    supporter.append(0)
  else:
    supporter.append(i)
del dataset['서포터 수']
dataset['서포터 수'] = supporter
```

```python
like1 = dataset['하트 수']
like = []
for i in like1:
  like.append(int(i))

del dataset['하트 수']
dataset['하트 수'] = like
```

```python
success1 = dataset['성공 여부']
success = []
for i in success1:
  j = i.split('%')
  j = int(j[0])
  success.append(j)

del dataset['성공 여부']
dataset['성공 여부'] = success
```

```python
feedback1 = dataset['메이커 명']
feedback = []
for i in feedback1:
  j = re.sub(r'[^0-9]', '', i)
  if j == '':
    feedback.append(100)
  else:
    feedback.append(j)

dataset['반응률'] = feedback
```
