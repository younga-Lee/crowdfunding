```python
def clean_text(inputString):
  text_rmv = re.sub('[-=✣+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\>\[\]`\'…》\”\“\’·]', ' ', inputString)
  text_rmv = ' '.join(text_rmv.split())
  return text_rmv
```

```python
stopwords = ['개발','배송','제품','사용','문의','발송','구조','설치','결제','제공','경우','리워드','인증','완료','혜택','와디','성능','서포터','가격','프로젝트','얼리버드','가능','등록','설정','한국','진행','시리즈','국내','생산','정책','만족도']
for i in list3:
  for j in i:
    if j in stopwords:
      i.remove(j)
```

```python
from collections import Counter
search_word = []
for i in list3:
  fre = []
  for j in i:
    fre.append(j)

  #가장 많이 나온 단어 상위 10개
  fre2 = list(Counter(fre).most_common(10))

  #길이별로 많은 단어 10개 넣은거 리스트에
  fre4 = []
  for j in fre2:
    fre4.append(list(j))
  fre4.sort(key=lambda x:len(x[0]), reverse = True)

  #길이가 2 초과인 단어들 먼저 넣고, num이라고 하는애가 초기값이야.
  #길이가 2인애들이 하나 추가될때마다 1씩 증가해서, num이 2가되면 안넣게
  #길이가 2 초과인 애들은 넣어지고, 2인애들은 2개씩만
  fre3 = []
  num = 0
  for k in fre4:
    if len(k[0]) > 2:
      fre3.append(k[0])
    else:
      if num == 2:
        break
      else:
        num = num+1
        fre3.append(k[0])

  #3중리스를 2중리스트로 바꾼거야
  fre3 = ' '.join(fre3)
  


  search_word.append(fre3)
```
